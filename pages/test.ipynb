```python
# df_손익_gr2 = df_손익[df_손익['중분류']=="매출"].groupby(by=['회계연도','전기월','보고반영']).sum()[['금액']]*-1
# df_손익_gr2 = df_손익[df_손익['중분류']=="매출"].groupby(by=['회계연도','보고반영']).sum()[['금액']]*-1
df_손익_gr2 = df_손익[df_손익['중분류']=="매출"].groupby(by=['회계연도']).sum()[['금액']]*-1
df_손익_gr2 = df_손익_gr2.reset_index()
df_손익_gr2 = df_손익_gr2[df_손익_gr2['회계연도'] =='2017']
df_손익_gr2
# df_손익_gr2.head()
```

```python
df_손익_gr2 = df_손익[df_손익['중분류']=="매출"].groupby(by=['회계연도','보고반영','전기월']).sum()[['금액']]*-1/1000000
df_손익_gr2 = df_손익_gr2.reset_index()
targets =['2023','2024']
df_손익_gr2 = df_손익_gr2.loc[(df_손익_gr2['회계연도'].isin(targets)) & (df_손익_gr2['전기월']<=7)]
df_손익_gr2
```

```python
df_손익_gr2 = df_손익_gr2.groupby(['회계연도','보고반영'])['금액'].agg(sum).reset_index()
df_손익_gr2 = df_손익_gr2.sort_values('금액',ascending=False)
df_손익_gr2
```

```python
fig = px.bar(df_손익_gr2, x= "회계연도", y="금액",  title='')
st.plotly_chart(fig, use_container_width=True)
```
